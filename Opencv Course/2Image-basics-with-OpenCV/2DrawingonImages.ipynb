```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.ones(shape=(512,512,3),dtype=np.int16)
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(380,10),pt2=(510,150),color=(0,255,0),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(255,150,100),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img,center=(100,100),radius=50,color=(0,0,255),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img,center=(25,25),radius=50,color=(0,0,255),thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img,pt1=(0,0),pt2=(512,512),color=(102,255,255),thickness=5)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text="Text",org=(10,500),fontFace=font,fontScale = 4,color=(255,255,255),thickness = 3,lineType= cv2.LINE_AA)
plt.imshow(blank_img)
```
